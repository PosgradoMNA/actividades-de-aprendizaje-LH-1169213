# Tema 1: Numpy 1D arrays

## Características

*   Es similar a una lista
*   Tiene un tamaño fijo
*   Cada elemento es del mismo tipo

```
# Cast the list and access via index

import numpy as np

a = np.array([0,1,2,3,4)]

a[0] #El resultado que nos dará es: 0

# Para revisar el tipo de dato se realiza lo siguiente

type(a): numpy.ndarray

a.dtype # El resultado es dtype("int64") por ser intergers y si fuera float, sería dtype("float64")

# Para obtener el tamaño 

a.size # El resultado es 5

# Número de dimensiones

a.ndim # El resultado es 1

# Tuple of intergers indicating the size of the array

a.shape # El resultado es (5,)

```


## Index and slicing

```
c = np.array ([20,1,2,3,4)]

# Para cambiar el primer elemento del array 

c[0]=100 # El resultado va a ser ([100,1,2,3,4)]

# Para cambiar el cuarto elemento del array

c[4]=0 # El resultado va a ser ([100,1,2,3,0)]

# Para cambiar un conjunto de elementos del array

c[3:5]=300,400 # El resultado va a ser ([100,1,2,300,400)]

# Para hacer un nuevo array del que ya creamos, se hace:

d = c[1:4] # El resultado va a ser ([1,2,3,)]

```
