## Operar con el archivo

File = open("file_path_and_name.txt",**mode**)

Common **modes** (sólo se usa la letra):

*   r -> reading
*   w -> writing
*   a -> appending


```
File1 = open("file_path_and_name.txt",w)
```

**File1.read()** -> para leer el archivo

**File1.name()**-> String with name of file

**File1.mode()** -> "r" is shown representing read

**File1.close()** -> para cerrar el documento; se tiene que hacer cada vez que se abre // *alternativa a continuación para que sea automático*

## Código para abrir archivo, realizar operaciones y cerrarlo en automático.

```
# Alternativa para cerrar automáticamente el programa

with open("file_path_and_name.txt","r" as File1:
  #Operations that will be run:
  file_stuff=File1.read() # Stores the content of file in the variable file_stuff
  print(file_stuff)
#End of the block, close file object:
print(File1.closed)
print(file_stuff) # Esto va a imprimir cada línea de resultado, si se examina, traerá el resultado con \n que indica el salto de línea

```

> This is line 1
  
>  This is line 2

Para conocer el **contenido del archivo por línea** se realiza lo siguiente:

```
with open("file_path_and_name.txt","r") as File1

```

> file_stuff: ["This is line 1\n", "This is line 2 \n", "This is line 3"]
